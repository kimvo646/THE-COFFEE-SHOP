```python
import pandas as pd

# Đường dẫn đến file Excel của bạn
file_path = '/content/sales reciepts raw_data(descriptive_statistics).xlsx'

# Đọc dữ liệu từ sheet tên 'cate'
df = pd.read_excel(file_path, sheet_name='Cate')

# In ra 5 dòng đầu tiên của dữ liệu
print(df.head())
```

```python
# Sử dụng io.BytesIO để đọc nội dung file
df = pd.read_csv(io.BytesIO(uploaded[filename]), encoding='latin1')
df
```

```python
import matplotlib.pyplot as plt

# Vẽ bar chart cho mỗi biến trong df
for column in df.columns:
    plt.figure(figsize=(10, 5))  # Tạo figure mới với kích thước tùy chỉnh
    df[column].value_counts().plot(kind='bar')  # Vẽ bar chart
    plt.title(column)  # Đặt tiêu đề cho biểu đồ
    plt.show()  # Hiển thị biểu đồ
```

```python
import matplotlib.pyplot as plt

# Vẽ pie chart cho mỗi cột trong df
for column in df.columns:
    plt.figure(figsize=(10, 5))  # Tạo figure mới với kích thước tùy chỉnh
    df[column].value_counts().plot(kind='pie', autopct='%1.1f%%')  # Vẽ pie chart
    plt.title(column)  # Đặt tiêu đề cho biểu đồ
    plt.ylabel('')  # Xóa nhãn trục y
    plt.show()  # Hiển thị biểu đồ
```
